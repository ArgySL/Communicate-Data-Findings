> Of the aforementioned features, five(5) are descriptive of Pokemon characteristics(type_1, speed, attack, agility_level, total), while "legendary" and "generation" define certain statuses to the Pokemon. To elaborate on that, "legendary" is a status that defines Pokemon that are unique, as in only one count of each speciment is supposed to exist, whereas "generation" categorizes Pokemon based on their order of Pokemon game chronological release.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('Pokemon_clean.csv')
```

## Bubbles or Tackles

In other words, Special Attack or Attack? This part of the presentation is focused on whether a single-Pokemon trainer (Josh, of course!) should train a pokemon that specializes in dealing damage either via special attacks or attacks.

```python
sb.set_style("whitegrid")
fg1 = sb.FacetGrid(data = df, hue = 'legendary', height = 7,
                 palette = 'colorblind')
fg1.map(plt.scatter, 'special_attack', 'attack')
fg1.set_axis_labels("Special Attack","Attack")
fg1.add_legend();
fg1.fig.suptitle('Comparison of Attackers and Special Attackers');
```

It is not very clear which kind of offensive Pokemon has the edge; in fact, it appears that Attacker Pokemon have the edge over Special Attackers, however...

What if Josh wanted to factor in a Pokemon's Speed characteristic as well? As it is widely known, in turn-based games,
the player who goes first has an advantage over the other player. What if a pokemon's Speed were to be factored in, when
making the decision, as well?


```python
fg2 = sb.FacetGrid(data = df, hue = 'agility_level', height = 7,
                 palette = 'colorblind')
fg2.map(plt.scatter, 'attack', 'speed',)
fg2.set_axis_labels("Attack","Speed")
fg2.add_legend();
fg2.fig.suptitle('When factoring in speed as well...');
```

```python
fg3 = sb.FacetGrid(data = df, hue = 'agility_level', height = 7,
                 palette = 'colorblind')
fg3.map(plt.scatter, 'special_attack', 'speed',)
fg3.set_axis_labels("Special Attack","Speed")
fg3.add_legend();
fg3.fig.suptitle('');
```

As it is visible in the upper-right end of the two graphs, **five(5)** pokemon labelled as "Very fast" for Attackers, versus **seven(7)** pokemon labelled as "Very Fast" for Special Attackers make their appearance, while *none* labelled as "Extremely Fast" for either of them (only Pokemon with either over 150 attack or special attack are included)!

Based on the results, it is safe to conclude that Josh should pick a **Special Attacker** Pokemon to train when prioritizing Speed!

## Bonus: Legendary Pokemon Power Across Generations
>Every guy or girl who has played Pokemon in their lifetime dreamed of catching and training Legendary pokemon, at least all my friends did! The visualization that follows showcases the power distribution of legendary Pokemon across generations.


```python
fg3 = sb.FacetGrid(data = df, hue = 'legendary', size = 7)
fg3.map(plt.scatter, 'total', 'generation')
fg3.add_legend();
```

Legendary Pokemon made a really strong debut with the first generation, however their powercreep settled down with the second generation, only to become much stronger on the third! The following generations' legendaries were toned down a bit, however they still remained overall strong, compared to their non-legendary counterparts.

To conclude this bonus part, one insight that can be derived from this graph is that the strongest legendary Pokemon belong in the first and third generations, so if you want to catch and train the very best, you should play Pokemon games that include pokemon from these generations!
